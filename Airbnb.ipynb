### Hotel Details Part

```python
#def Hotel_Details():
airbnb_hotel_data = []

for item in data:
    Hotel_data = {
            "Hotel_Id": item.get("_id"),
            "Hotel_Name": item.get("name"),
            "Hotel_url": item.get("listing_url"),
            "Description": item.get("description"),
            "Property_type" : item.get("property_type"),
            "Room_type" : item.get("room_type"),
            "Bed_type": item.get("bed_type"),
            "Accomodates": item.get("accommodates"),
            "Min_nights": item.get("minimum_nights"),
            "Price": item.get("price"),
            "Availability_365": item.get("availability", {}).get("availability_365"),
            "Amenities": item.get("amenities"),
            "Neighbourhood" : item.get("address",{}).get("government_area"),
            "Neighbourhood_group" : item.get("address",{}).get("suburb"),
            "City" : item.get("address",{}).get("market"),
            "Country" : item.get("address",{}).get("country"),
            "latitude": item.get("address", {}).get("location", {}).get("coordinates")[1],
            "longitude": item.get("address", {}).get("location", {}).get("coordinates")[0],
            "Host_Id": item.get("host", {}).get("host_id"),
            "Host_Name": item.get("host", {}).get("host_name"),
            "Host_url": item.get("host",{}).get("host_url"),
            "Total_listing_by_Host" : item.get("host",{}).get("host_total_listings_count"),
            "Num_of_review" : item.get("number_of_reviews"),
            "Review_rating": item.get("review_scores", {}).get("review_scores_rating")}
            
        
    airbnb_hotel_data.append(Hotel_data)
Airbnb_Hotel_df = pd.DataFrame(airbnb_hotel_data)

# return Airbnb_Hotel_df
```

```python
def find_non_hashable_columns(df):
    non_hashable_columns = []
    for column in df.columns:
        if df[column].apply(lambda x: isinstance(x, (list, dict, set, bytearray))).any():
            non_hashable_columns.append(column)
    return non_hashable_columns

non_hashable_columns = find_non_hashable_columns(Airbnb_Hotel_df)

# Convert Non-Hashable to Hashable

for column in non_hashable_columns:
    # Convert lists to tuples
    Airbnb_Hotel_df["Amenities"] = Airbnb_Hotel_df["Amenities"].apply(
        lambda x: tuple(x) if isinstance(x, list) else x
    )
```

### Hotel Review Part

```python
#def Review_Details():
airbnb_review_data = []

for item in data:
    reviews = item.get("reviews", [])

    for review in reviews:
        Review_data = {
            "Hotel_Id": item.get("_id"),
            "Hotel_Name": item.get("name"),
            "Reviewer_Id": review.get("reviewer_id"),  
            "Reviewer_Name": review.get("reviewer_name"),
            "Comments": review.get("comments")
        }
        airbnb_review_data.append(Review_data)
    Airbnb_Review_df = pd.DataFrame(airbnb_review_data)

#return Airbnb_Review_df
```

```python
Airbnb_Review_df = Airbnb_Review_df[Airbnb_Review_df["Comments"] != ""]
Airbnb_Review_df = Airbnb_Review_df[Airbnb_Review_df["Comments"] != "-"]
Airbnb_Review_df = Airbnb_Review_df[Airbnb_Review_df["Comments"] != "---"]
Airbnb_Review_df = Airbnb_Review_df[Airbnb_Review_df["Reviewer_Name"] != "(Email hidden by Airbnb)"]
Airbnb_Review_df = Airbnb_Review_df[Airbnb_Review_df["Reviewer_Name"] != "(Website hidden by Airbnb)"]
Airbnb_Review_df = Airbnb_Review_df[Airbnb_Review_df["Reviewer_Name"] != "(Phone number hidden by Airbnb)"]
```
